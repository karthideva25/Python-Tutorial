### Load Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

# Resize your graph
plt.figure(figsize=(5,3), dpi=250)

# plot or mark value
#plt.plot(x,y, label='2x', color="green", linewidth=3, marker=".", linestyle="--" ,markersize=10, markeredgecolor="red")

#use short-hand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--' ,label='2x')

# Line number two
x2 = np.arange(0,4,0.5)
print(x2)
#plt.plot(x2,x2**2, 'r',label='X')

#We need first four values
# plt.plot(x2[:4],x2[:4]**2, 'r',label='X')
plt.plot(x2[:6],x2[:6]**2, 'r', label="x^2")
plt.plot(x2[5:],x2[5:]**2, 'r--')

# title for graph
plt.title("Our first Graph",fontdict={"fontname":"Comic Sans Ms","fontsize":20})

#lable x axis
plt.xlabel("X axis")

#lable y axis
plt.ylabel("Y axis")

# give our own x axis value
plt.xticks([1,2,3,4])

# give our own y axis value
plt.yticks([2,4,6,8,10])

# using lable to mark legend
plt.legend()

#Save Graph
plt.savefig('Mygraph.png', dpi=350)

# show graph
plt.show()
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

# plot bar graph
bars = plt.bar(labels,values)

# shade bar graph
patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

### Real world Example

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5),dpi=300)

plt.title("Gas prices over Time (in USD)", fontdict={"fontweight":'bold',"fontsize":18})

plt.plot(gas.Year,gas.USA, 'b.-',label="USA" )
plt.plot(gas.Year,gas.Canada, 'r.-',label="Canada")
plt.plot(gas.Year,gas['South Korea'], 'g.-',label="South Korea")
plt.plot(gas.Year,gas.Australia, 'y.-', label="Australia")

# Another way to plot many values


# countries_to_look_at = ['Australia','USA','Canada','South Korea']
# for country in gas:
# #     if country != "Year":
#     if country in countries_to_look_at:
#         plt.plot(gas["Year"],gas[country],marker='.',label=country)

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel("Years")
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('gas_price.png',dpi=300)

plt.show()
```

### Load Fifa data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall,bins=bins,color='#abcdef')

plt.xticks(bins)

plt.xlabel('Skill level')
plt.ylabel("Number of players")
plt.title("Distribution of player skills in FIFA 2018")

plt.show()
```

### Pie chart

```python
left = fifa.loc[fifa["Preferred Foot"] == 'Left'].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == 'Right'].count()[0]

labels = ['Right','Left']
colors = ['#abcdef','#aabbcc']

plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')

plt.title("Foot preference of FIFA players")
# plt.legend()

plt.show()
```

### Another pie chart

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

plt.figure(figsize=(5,5), dpi=300)

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125)&(fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150)&(fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175)&(fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ["Under 125", '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.2,0,0,.4)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.savefig("piechart.png",dpi=400)

plt.show()
```
